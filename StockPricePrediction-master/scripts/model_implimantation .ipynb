```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('FB.csv')
columns = dataframe.columns
adjclose = columns[-2]
returns = columns[-1]
maxdelta = 30
delta = range(8,maxdelta)
dataframe.head()
```

```python
for n in delta:
    return_n = adjclose[9:] + "Time" + str(n)
    dataframe[return_n] = dataframe[adjclose].pct_change(n)
    
    roll_n = returns[7:] + "RolMean" + str(n)
    dataframe[roll_n] = pd.rolling_mean(dataframe[returns], n)

    exp_ma = returns[7:] + "ExponentMovingAvg" + str(n)
    dataframe[exp_ma] = pd.ewma(dataframe[returns], halflife=n)
dataframe = dataframe.iloc[max(delta):,:]
dataframe.head()
```

```python
from datetime import datetime
if 'Date' in dataframe:
        dataframe['new_col'] = pd.to_datetime(dataframe['Date']).astype(datetime)
        #file_dataframe['date'] = pd.to_datetime(file_dataframe['date'])
        dataframe['new_col'].apply(lambda dt_time:10000*dt_time.year + 1000*dt_time.month + dt_time.day).astype(int)
```
